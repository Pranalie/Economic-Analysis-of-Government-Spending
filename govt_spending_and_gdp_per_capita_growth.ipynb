**Government spending and gdp per capita growth**

Load libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_excel('GDP per capita growth (annual %.xls', skiprows=[0, 1,2])
data1 = pd.read_excel('General government final consumption expenditure (current US$).xls', skiprows=[0, 1,2])
```

Choose G20 Nations

```python
specified_countries = [
    "Argentina", "Australia", "Brazil", "Canada", "China", "France", "Germany", "India",
    "Indonesia", "Italy", "Japan", "Korea, Rep.", "Mexico", "Russian Federation",
    "Saudi Arabia", "South Africa", "Turkiye", "United Kingdom", "United States",
    "European Union"
]

filtered_data = data[data['Country Name'].isin(specified_countries)]
filtered_data1 = data1[data1['Country Name'].isin(specified_countries)]
```

```python
filtered_data = filtered_data.drop('Indicator Name', axis=1)
filtered_data1 = filtered_data1.drop('Indicator Name', axis=1)
```

```python
filtered_data = pd.melt(filtered_data, id_vars=['Country Name', 'Country Code', 'Indicator Code'], value_vars=['1960', '1961',
       '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970',
       '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979',
       '1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988',
       '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997',
       '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006',
       '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015',
       '2016', '2017', '2018', '2019', '2020', '2021', '2022'],
        var_name='Year', value_name='Value')

filtered_data1 = pd.melt(filtered_data1, id_vars=['Country Name', 'Country Code', 'Indicator Code'], value_vars=['1960', '1961',
       '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970',
       '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979',
       '1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988',
       '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997',
       '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006',
       '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015',
       '2016', '2017', '2018', '2019', '2020', '2021', '2022'],
        var_name='Year', value_name='Value')
```

Let's see which country had the highest GDP per capita growth in 2022.

```python
filtered_data = filtered_data.dropna()
filtered_data_ = filtered_data[filtered_data['Year']  == '2022']
filtered_data_ = filtered_data_.sort_values(by='Value', ascending=False)
#filtered_data_
```

**GDP Per Capita Growth Over Time - India vs. China vs. United States from 1961 to 2022**

```python
new_filtered_data = filtered_data[filtered_data['Country Name'].isin(['India','China','United States'])]
pivoted_data = new_filtered_data.pivot(index='Year', columns='Country Name', values='Value')
pivoted_data = pivoted_data.dropna()

# Create a time series plot with Seaborn
plt.figure(figsize=(12, 6))

# Define a custom color palette
colors = ['blue', 'orange','green']

# Use the palette for color consistency
sns.set_palette(colors)

sns.lineplot(x=pivoted_data.index, y=pivoted_data['China'], marker='o', markersize=7, linestyle='-', label='China')
sns.lineplot(x=pivoted_data.index, y=pivoted_data['India'], marker='o', markersize=7, linestyle='-', label='India')
sns.lineplot(x=pivoted_data.index, y=pivoted_data['United States'], marker='o', markersize=7, linestyle='-', label='United States')

# Add labels and a title
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP per capita growth (annual %)', fontsize=12)
plt.title('GDP Per Capita Growth Over Time', fontsize=16)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show only even years on the x-axis
even_years = pivoted_data.index[::2]
plt.xticks(even_years)

# Customize legend
plt.legend(title='Country', title_fontsize=12, loc='upper left')

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.7)

# Customize plot style
sns.set_style("whitegrid")

# Remove spines
sns.despine()

# Autoscale the y-axis to cover the entire range
plt.autoscale()

# Show the plot
plt.tight_layout()
plt.show()
```

**General government final consumption expenditure (current US$) Over Time**

```python
new_filtered_data1 = filtered_data1[filtered_data1['Country Name'].isin(['India','China','Korea, Rep.','Japan','Brazil'])]
pivoted_data1 = new_filtered_data1.pivot(index='Year', columns='Country Name', values='Value')
pivoted_data1 = pivoted_data1.dropna()

# Create a time series plot with Seaborn
plt.figure(figsize=(12, 6))

# Define a custom color palette
colors = ['blue', 'orange','purple','green','pink']

# Use the palette for color consistency
sns.set_palette(colors)

sns.lineplot(x=pivoted_data1.index, y=pivoted_data1['China'], marker='o', markersize=7, linestyle='-', label='China')
sns.lineplot(x=pivoted_data1.index, y=pivoted_data1['India'], marker='o', markersize=7, linestyle='-', label='India')
sns.lineplot(x=pivoted_data1.index, y=pivoted_data1['Korea, Rep.'], marker='o', markersize=7, linestyle='-', label='Korea, Rep.a')
sns.lineplot(x=pivoted_data1.index, y=pivoted_data1['Japan'], marker='o', markersize=7, linestyle='-', label='Japan')
sns.lineplot(x=pivoted_data1.index, y=pivoted_data1['Brazil'], marker='o', markersize=7, linestyle='-', label='Brazil')

# Add labels and a title
plt.xlabel('Year', fontsize=12)
plt.ylabel('General government final consumption expenditure (current US$)', fontsize=12)
plt.title('General government final consumption expenditure (current US$) Over Time', fontsize=16)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show only even years on the x-axis
even_years = pivoted_data1.index[::2]
plt.xticks(even_years)

# Customize legend
plt.legend(title='Country', title_fontsize=12, loc='upper left')

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.7)

# Customize plot style
sns.set_style("whitegrid")

# Remove spines
sns.despine()

# Autoscale the y-axis to cover the entire range
plt.autoscale()

# Show the plot
plt.tight_layout()
plt.show()
```
