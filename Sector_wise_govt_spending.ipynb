**Sector wise plot analysis**

In 2022, China led all G20 nations by making the highest investments in Agriculture, Forestry, and Fishing, Industry (including construction), and Manufacturing, value added, consistently holding the top position across all available years. Meanwhile, the United States secured its position at the forefront among G20 nations by investing the most in Services, value added in 2022 (these investment values are measured in current US dollars).

Load the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Read the files

```python
data1 = pd.read_excel('Agriculture, forestry, and fishing, value added (current US$).xls', skiprows=[0, 1,2])
data2 = pd.read_excel('Industry (including construction), value added (current US$).xls', skiprows=[0, 1,2])
data3 = pd.read_excel('Manufacturing, value added (current US$).xls', skiprows=[0, 1,2])
data4 = pd.read_excel('Services, value added (current US$).xls', skiprows=[0, 1,2])
```

Seperate G20 nations

```python
specified_countries = [
    "Argentina", "Australia", "Brazil", "Canada", "China", "France", "Germany", "India",
    "Indonesia", "Italy", "Japan", "Korea, Rep.", "Mexico", "Russian Federation",
    "Saudi Arabia", "South Africa", "Turkiye", "United Kingdom", "United States",
    "European Union"
]

filtered_data1 = data1[data1['Country Name'].isin(specified_countries)]
filtered_data2 = data2[data2['Country Name'].isin(specified_countries)]
filtered_data3 = data3[data3['Country Name'].isin(specified_countries)]
filtered_data4 = data4[data4['Country Name'].isin(specified_countries)]
```

```python
#Drop the column name Indicator Name
filtered_data1 = filtered_data1.drop('Indicator Name', axis=1)
filtered_data2 = filtered_data2.drop('Indicator Name', axis=1)
filtered_data3 = filtered_data3.drop('Indicator Name', axis=1)
filtered_data4 = filtered_data4.drop('Indicator Name', axis=1)
```

```python
#Rename the columns and merge them together.
filtered_data1.rename(columns={'Value': 'Agri, forestry, and fishing (current US$)'}, inplace=True)
filtered_data2.rename(columns={'Value': 'Industry (including construction) (current US$)$)'}, inplace=True)
filtered_data3.rename(columns={'Value': 'Manufacturing, value added (current US$)'}, inplace=True)
filtered_data4.rename(columns={'Value': 'Services, value added (current US$)'}, inplace=True)

result = pd.concat([filtered_data1,
                    filtered_data2['Industry (including construction) (current US$)$)'],
                    filtered_data3['Manufacturing, value added (current US$)'],
                    filtered_data4['Services, value added (current US$)']], axis=1)
result
```

**Agri, forestry, and fishing (current US$) - China vs India**

```python
#This shows that China leads the government spending in Agri, forestry, and fishing (current US$)followed by India in 2022.
sorted_df = result.sort_values(by='Agri, forestry, and fishing (current US$)', ascending=False)
sorted_df = sorted_df.reset_index(drop = True)
#sorted_df.head(20)
```

```python
year = ['1980', '1990', '2000', '2010', '2020']
new_filtered_data1 = filtered_data1[filtered_data1['Country Name'].isin(['India', 'China'])]
new_filtered_data1 = new_filtered_data1[new_filtered_data1['Year'].isin(year)]
#new_filtered_data1
```

```python
import matplotlib.pyplot as plt
agri_forestry_fishing_data = {
    'Year': [1980, 1990, 2000, 2010, 2020],
    'China': new_filtered_data1[new_filtered_data1['Country Name'] == 'China']['Agri, forestry, and fishing (current US$)'],
    'India': [6.411244e+10, 8.854171e+10, 1.012251e+11, 2.852988e+11, 4.978660e+11]
}

# Create a DataFrame from the data
import pandas as pd

df = pd.DataFrame(agri_forestry_fishing_data)

# Set 'Year' as the index for time series plotting
df.set_index('Year', inplace=True)

# Plot the time series
plt.figure(figsize=(10, 6))
plt.plot(df['China'], marker='o', label='China')
plt.plot(df['India'], marker='o', label='India')

plt.title('Time Series of Agri, Forestry, and Fishing (current US$)')
plt.xlabel('Year')
plt.ylabel('Agri, Forestry, and Fishing (current US$)')
plt.xticks(df.index)
plt.grid(True)
plt.legend()

plt.show()
```
