**Primary govt expenditures as a proportion of original approved budget (%)**

*Note : This dataset does not contain values before the year 2005. Also no data is available for countries such as United States and China. So, we'll restricting our attention to India.*

Load the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Read the file

```python
data = pd.read_excel('Primary government expenditures as a proportion of original approved budget (%).xls', skiprows=[0, 1,2])
data = data.drop('Indicator Name', axis=1)
```

Selecting the G20 nations

```python
specified_countries = [
    "Argentina", "Australia", "Brazil", "Canada", "China", "France", "Germany", "India",
    "Indonesia", "Italy", "Japan", "Korea, Rep.", "Mexico", "Russian Federation",
    "Saudi Arabia", "South Africa", "Turkiye", "United Kingdom", "United States",
    "European Union"
]

filtered_data = data[data['Country Name'].isin(specified_countries)]
filtered_data = filtered_data.dropna(axis=1, how='all')
```

```python
filtered_data = pd.melt(filtered_data, id_vars=['Country Name', 'Country Code', 'Indicator Code'], value_vars=['2005', '2006',
       '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021'],
        var_name='Year', value_name='Value')
```

Drop the Nan values from the dataset

```python
filtered_data = filtered_data.dropna()
```

Selecting India for further evaluation.

```python
new_filtered_data1 = filtered_data[filtered_data['Country Name'].isin(['India'])]
#new_filtered_data1
```

Plot using Matplotlib

```python
# Sample DataFrame with the provided data
data = { 'Year': new_filtered_data1['Year'],
    'Value1': new_filtered_data1['Value'],
}

df = pd.DataFrame(data)

# Create a bar plot (histogram)
plt.figure(figsize=(10, 6))
bars1 = plt.bar(df['Year'], df['Value1'], color='lightblue')

plt.xlabel('Year')
plt.ylabel('Primary govt expenditures as a proportion of original approved budget (%)')
plt.title('Bar plot (India)')
plt.xticks(rotation=45)
plt.grid(True)

# Label each bar with its value
for bar in bars1:
    height = bar.get_height()
    plt.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height), xytext=(0, 3),
                 textcoords='offset points', ha='center', va='bottom')

plt.show()
```
